```python
import pandas as pd
import numpy as np

data=pd.read_csv('/content/drive/MyDrive/Tableau_IMDB_Project/TMDB  IMDB Movies Dataset.csv')
```

```python
# Dropping of columns that are not required- status, backdrop_path, homepage, tconst, original_title, overview, poster_path, tagline

data1= data.drop(['status', 'backdrop_path', 'homepage', 'tconst', 'original_title', 'overview', 'poster_path', 'tagline'], axis=1)

data1
```

```python
# Converting release_date dtype from object to datetime

data1['release_date']=pd.to_datetime(data1['release_date'], errors='coerce')
```

```python
# Since the dataset is very large and has movies from 19th century as well, so lets filter out them to include movies having release year from 2000

data1= data1[data1['release_date'].dt.year>=2000]
```

```python
# Lets include movies having original language english

data2= data1[data1['original_language']=='en']
```

```python
data2
```

```python
# we will consider movies which are available for all age groups

data3= data2[data2['adult']==False]
```

```python
# sorting genre values of each row in alphabetical order to prevent it from duplication

def clean_genre(genres):
    if isinstance(genres, str):
      genre_list= genres.split(',')
      clean_list=[]
      for g in genre_list:
        clean_list.append(g.strip())
      clean_list.sort()
      return ' ,'.join(clean_list)

    else:
      return ''

new_column=[]
for value in data3['genres']:
    new_column.append(clean_genre(value))

data3['genres']=new_column
```

```python
# Saving the cleaned file

data3.to_csv('/content/drive/MyDrive/Tableau_IMDB_Project/TMDB+IMDB_New_Data.csv')
```
